## **Importing Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

## Loading dataset

```python
df0=pd.read_csv('C:/Data sets/batter.csv')
```

```python
df=sns.load_dataset('tips')
```

```python
df.head(5)
```

## **Scatter plot**

```python
sns.scatterplot(data=df,x='total_bill',y='tip',hue='sex',style='time')
plt.show()
```

## **Lineplot**

```python
gap=px.data.gapminder()
pak=gap[gap['country']=='Pakistan']
```

```python
plt.figure(figsize=(8,6),dpi=150)
sns.relplot(kind='line',data=pak,x='year',y='lifeExp')
plt.xlabel('year')
plt.ylabel('Lifeexp')
plt.show()
```

```python
con=gap[gap['country'].isin(['India','Germany','China'])]
```

```python
sns.relplot(kind='line',data=con,x='year',y='lifeExp',hue='country',style='continent')
```

## **Facet plot**

```python
sns.relplot(kind='scatter',data=df,x='total_bill',y='tip' ,col='sex')
```

```python
sns.relplot(kind='scatter',data=df,x='total_bill',y='tip' ,col='smoker',row='sex')
```

## **Bar plot**

```python
sns.barplot(data=df, x='time' , y='size',color='red')
plt.show()
```

```python
df.head()
```

## **Count plot**

```python
sns.countplot(data=df,x='smoker',hue='time')
```

##  **Pair plot**

```python
sns.pairplot(data=df,hue='sex',vars=['total_bill','tip'])
plt.xlabel('Total bill')
plt.ylabel('Tip')
plt.show()
```

```python
sns.catplot(data=df,x='day',y='total_bill',hue='sex',kind='box')
plt.title('Box plot of total bill by day')
plt.xlabel('Day of the week')
plt.ylabel('Total bill amount')
plt.show()
```

```python
sns.boxenplot(data=df,x='day',y='total_bill',hue='sex')
plt.title('Boxen plot of total bill by day')
plt.xlabel('Day of the week')
plt.ylabel('Total bill amount')
plt.show()
```

```python
sns.displot(data=df,x='total_bill',hue='sex',kde=True)
```

```python
num = df.select_dtypes(include=[np.number])
corr = num.corr()
plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation matrix (numeric columns)')
plt.show()
```
